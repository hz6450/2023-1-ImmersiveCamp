```python
from bs4 import BeautifulSoup
```

```python
ex = '''
    <html>
        <head>
            <title> 김밥천국에 점심을 묻다 </title>
        </head>
        <body>
        <p align="left"> 치즈김밥 </p>
        <p align="center"> 멸치김밥 </p>
        <p align="right"> 샐러드김밥 </p>
    </body>
    </html>
'''

soup = BeautifulSoup(ex,'html.parser')
#soup.find('p', align="right")
soup.find_all('p')
```

```python
txt = soup.find_all('p')

for i in txt:
    print(i.get_text())
    
soup.select('p')
```

```python
import requests
from bs4 import BeautifulSoup
url = "https://datalab.naver.com"  
response = requests.get(url)   #resqonse에 url주소 저장

soup = BeautifulSoup(response.text,'html.parser')
soup
pitems = soup.select('span.title')  #span태그안의 title이 있는 데이터 가져오기
pitems

for i in pitems:
    print(i.get_text())
```

```python
import re

pattern = re.compile("[가-힣]+")  # Pattern to match Korean characters at the end
titles = [pattern.findall(str(pitem))[0] if pattern.findall(str(pitem)) else '' for pitem in pitems]
print(titles)

```

```python
import pandas as pd

df = pd.DataFrame({'titles': titles})  # Use 'titles' as the column name
df.to_excel('titles.xlsx', index=False)  # Corrected file name
df.to_csv('titles.csv', index=False)  # Corrected file name
```
